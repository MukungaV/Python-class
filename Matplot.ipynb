```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('X vs Y')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])

axes1.plot(x,y,marker='s')

axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('X vs Y')


axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes2.plot(x,y,color='red',marker='o')

axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('X vs Y')
```

```python
fig=plt.figure(figsize=(12,6))
axes1=fig.add_axes([0.1,0.1,0.8,0.8])

axes1.plot(x,x+1,lw=5)
axes1.plot(x,x+2,ls='--')
axes1.plot(x,x+3,ls=':')
axes1.plot(x,x+4,ls='-')

```
